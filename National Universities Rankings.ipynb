# Read Data 

```python
data=pd.read_csv('National Universities Rankings.csv')
```

```python
data.info()
```

Columns
File: National Universities Rankings.csv

**Name	The name of the university. (String)** 

**Location	The city and state where the university is located. (String)** 


**Rank	The ranking of the university according to U.S News & World Report. (Integer)**


**Description	A snippet overview from US News. (String)**


**Tuition and fees	Combined tuition fee rate for out-of-state students. (Integer)**


**In-state	Combined tuition for in-state students. (Integer)**


**Undergrad Enrollment	Number of enrolled undergraduates at the school. (Integer)**

```python
data.drop('index',axis=1,inplace=True)
```

```python
data.head(5)
```

```python
def clean(text):
    text=text.replace('$','')
    text=text.replace(',','')
    return text
```

```python
data['Tuition and fees']=data['Tuition and fees'].apply(clean)
```

```python
data['Undergrad Enrollment']=data['Undergrad Enrollment'].apply(clean)
```

```python
data['Tuition and fees']=data['Tuition and fees'].astype('int64')
```

```python
data['Undergrad Enrollment']=data['Undergrad Enrollment'].astype('int64')
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop('Description',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.drop('In-state',axis=1,inplace=True)
```

```python
data.head()
```

```python
def extract_state(text):
    return text[-2:]
```

```python
data['states']=data['Location'].apply(extract_state)
```
